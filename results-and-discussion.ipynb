```python
def small_table(year,big_df,yvar):
    small_df = big_df.loc[:,['Country Code',str(year)+'_pop',str(year)+'_'+str(yvar)]]
    small_df.dropna(inplace = True)
    return small_df
```

```python
def plot_var_year_single_plots (df,yvar, yname,year,betas,errors=[], xmin=None, xmax=None, ymin=None, ymax=None):
    t = small_table(year, df, yvar)
    x = t[str(year)+"_pop"].apply(np.log)
    y = t[str(year)+"_"+str(yvar)].apply(np.log) 
    
    if len(t) == 0:
        return
    
    model = sm.OLS(y,sm.add_constant(x))
    results = model.fit()
    betas.append(results.params[1])
    errors.append(results.bse[1])

    if year in (1961,1971,1981,1991,2001,2011): # 1965,1975,1985,1995,2005):   
        plt.scatter(x=x, y=y, marker='o', color='k', alpha = 0.8)
        plt.xlabel("Log Total Population")
        plt.ylabel("Log "+ str(yname))
        fit_line = np.linspace(x.min(), x.max(), 1000)
        plt.plot(fit_line, fit_line*results.params[1] + results.params[0], color='blue', linewidth = 1.5, alpha = 0.8)
        title = "Countries Population Vs. "+str(yname) + " For "+str(year)
        plt.title(title)
        plt.text(x.min(),y.max(),"Slope = "+str(results.params[1]))
        plt.text(x.min(),y.max()-1,"R Squared = "+str(results.rsquared))
        if not xmin:
            xmin =  x.min()-1
        if not xmax:
            xmax = x.max()+1
        if not ymin:
            ymin = y.min()-1
        if not ymax:
            ymax = y.max()+1            
        plt.axis([xmin,xmax,ymin,ymax])
        filename = title.replace(" ","_")
        plt.gcf().savefig("output/" + filename + ".jpg")

        plt.show()
        print results.summary()
```

```python
patent = pd.read_csv("data/patent.csv")
patent_non = pd.read_csv("data/patent_non.csv")
patent_comb = pd.DataFrame()
patent_comb['Country Code'] = patent['Country Code']
for y in range(1961,2013):
    patent_comb[str(y)] = patent[str(y)] + patent_non[str(y)]
patent_comb["Country Name"] = patent.ix[:,0] # patent["Country Name"]
patent_comb.tail()
```

```python
# pop was already read in at top of file
carbon = pd.read_csv("data/co2.csv")
gdp = pd.read_csv("data/gdp.csv")
hom = pd.read_csv("data/homicides.csv")

remove_bom(carbon)
remove_bom(gdp)
remove_bom(hom)

pop.tail()
```

```python
carbon_pop = pd.merge(pop, carbon, on = ['Country Code'], how = 'inner', suffixes=['_pop', '_carbon'])
gdp_pop = pd.merge(pop, gdp, on = ['Country Code'], how = 'inner', suffixes=['_pop', '_gdp'])
patent_pop = pd.merge(pop, patent_comb, on = ['Country Code'], how = 'inner', suffixes=['_pop', '_patent'])
hom_pop = pd.merge(pop, hom, on = ['Country Code'], how = 'inner', suffixes=['_pop', '_hom'])
for y in range(1961,2014):
    hom_pop[str(y)+"_hom"] = (hom_pop[str(y)+"_hom"] * hom_pop[str(y)+"_pop"]) / 100000

print "There were %d lines of data on carbon. We combined it %d of them with population data" % ( len(carbon), len(carbon_pop) )
print "There were %d lines of data on patents. We combined it %d of them with population data" % ( len(patent), len(patent_pop) ) 
```

```python
def plot_var_year (df,yvar, yname,year,betas,errors=[], summaries=[], xmin=None, xmax=None, ymin=None, ymax=None, ax=None):
    t = small_table(year, df, yvar)
    x = t[str(year)+"_pop"].apply(np.log)
    y = t[str(year)+"_"+str(yvar)].apply(np.log) 
    
    if len(t) == 0:
        print "No data in %d, ignoring" % year
        return
    
    model = sm.OLS(y,sm.add_constant(x))
    results = model.fit()
    betas.append(results.params[1])
    errors.append(results.bse[1])

    if ax:
        summaries.append(results.summary())
        ax.scatter(x=x, y=y, marker='o', color='k', alpha = 0.8)
        fit_line = np.linspace(x.min(), x.max(), 1000)
        ax.plot(fit_line, fit_line*results.params[1] + results.params[0], color='blue', linewidth = 1.5, alpha = 0.8)
        title = " Year "+str(year)
        ax.set_title(title)
   
        if not xmin:
            xmin =  x.min()-1
        if not xmax:
            xmax = x.max()+1
        if not ymin:
            ymin = y.min()-1
        if not ymax:
            ymax = y.max()+1            
        ax.axis([xmin,xmax,ymin,ymax])
        ax.text(xmin+1,ymax-2,"Scaling Exponent = %2.2f" % results.params[1])
        ax.text(xmin+1,ymax-4,"R Squared = %2.2f" % results.rsquared)
        # if (len(y)<15):
        ax.text(xmin+1,ymax-6,"n = %d" % len(x))
            

        
```

```python

fig, ax = plt.subplots(1, 1)
i=0
summaries = []

plot_var_year (carbon_pop,"carbon","Carbon Emission",1961, [], errors = [], summaries=summaries, xmin=8, xmax=23, ymax=18, ax=ax)


ax.set_ylabel("Log Carbon") 
ax.set_xlabel("Log Total Population")

plt.suptitle("Carbon Emission vs. Population", fontsize=16)
fig.savefig("output/carbon_emission_1961.jpg")
plt.show()

print summaries

with open('output/summary-carbon-1961.txt', 'w') as the_file:
    the_file.write(summaries[0].as_text())
    the_file.close()
```

```python
betas = []
errors = []
summaries = []
years = range(1961,2011)
fig, axes = plt.subplots(3, 2, sharex='col', sharey='row')
axes = axes.flatten()
i=0

for year in years:
    if year in (1961,1970,1980,1990,2000,2010): # 1965,1975,1985,1995,2005):   
        ax = axes[i]
        print "plot %d on ax%d" % (year, i)
        i += 1
    else:
        ax = None
    plot_var_year (carbon_pop,"carbon","Carbon Emission",year, betas, errors = errors, summaries=summaries, xmin=8, xmax=23, ymax=18, ax=ax)


axes[0].set_ylabel("Log Carbon") 
axes[2].set_ylabel("Log Carbon") 
axes[4].set_ylabel("Log Carbon") 
axes[4].set_xlabel("Log Total Population")
axes[5].set_xlabel("Log Total Population")
print "my betas are:"
print betas
print "my errors are:"
print errors

plt.suptitle("Carbon Emission vs. Population", fontsize=16)
fig.savefig("output/carbon_emission.jpg")
plt.show()
```

```python
print "got %d values and %d errors for %d years" % ( len(betas), len(errors), len(years) )
#plt.plot(years, betas, marker = 'o', color = 'k', alpha = 0.8)

plt.errorbar(years, betas, marker = 'o', color = 'k', alpha = 0.8, yerr=errors)
 
plt.plot(years, np.ones(len(years)), color='red', linewidth = 1.5)
# ax0.set_aspect('equal', 'datalim')

plt.axis([min(years)-1,max(years)+1,min(betas)-max(errors),max(betas)+max(errors)])
title = "Carbon Emission - Scaling Exponents Over Time"
plt.title(title)
plt.xlabel("Year")
plt.ylabel("Scaling Exponent")
filename = title.replace(" ","_")
plt.gcf().savefig("output/" + filename + ".jpg")

plt.show()
```
